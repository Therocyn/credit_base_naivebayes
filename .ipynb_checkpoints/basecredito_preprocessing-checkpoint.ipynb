# **29- Divisão de base de treino e base de teste:**

```python
from sklearn.model_selection import train_test_split
```

```python
X_credit_treinamento, X_credit_teste, y_credit_treinamento, y_credit_teste = train_test_split(X_credit, y_credit, test_size=0.25, random_state=0) #split pra base de dados de crédito
```
